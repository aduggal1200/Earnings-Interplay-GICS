```python

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_heatmap(data_dict):
    """
    Plot a heatmap based on a dictionary of data.

    Parameters:
    - data_dict (dict): Dictionary containing the data for the heatmap.

    Returns:
    - None
    """
    # Convert dictionary data to DataFrame
    data_df = pd.DataFrame(list(data_dict.values()),
                           index=data_dict.keys(),
                           columns=['Average'])

    # Plotting heatmap
    plt.figure(figsize=(15, 5))
    sns.heatmap(data_df.T, cmap='coolwarm', annot=True,
                linewidths=0.5, cbar_kws={'label': 'Average'})
    plt.xticks(ha='center')
    plt.yticks(rotation=0, fontsize=10)
    plt.title('Average Correlation for the Whole Sector')
    plt.show()


# Data
data = {
    'Utilities': 0.2495166667,
    'Real Estate': 0.1952428571,
    'Materials': 0.04335,
    'IT': 0.1411,
    'Industrials': 0.24825,
    'Health Care': 0.19265,
    'Financials': 0.12098,
    'Energy': 0.269975,
    'Consumer Staples': 0.111025,
    'Consumer Discretionary': 0.27617,
    'Communication Services': 0.17473
}

# Plot
plot_heatmap(data)
```
